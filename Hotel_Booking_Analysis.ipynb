### Check Unique Values for each variable.

````python
# Check Unique Values for each variable.
def var(hotel_df):
    unique_list = pd.DataFrame([[i,len(hotel_df[i].unique())] for i in hotel_df.columns])
    unique_list.columns = ['name','uniques']

    total_var = set(hotel_df.columns)
    cat_var = set(unique_list.name[(unique_list.uniques<=12)      |
                                   (unique_list.name=='Country')  |
                                   (unique_list.name=='Agent')
                                  ])
    con_var = total_var - cat_var

    return cat_var, con_var


cat_var, con_var = var(hotel_df)

print("Continuous Variables (",len(con_var),")\n",con_var,'\n\n'
      "Categorical Variables(",len(cat_var),")\n",cat_var)
````

# **Conclusion**

In conclusion, our analysis reveals key insights that should inform our strategic approach to hotel bookings and advertising. The majority of our booked hotels belong to the city category, indicating a strong preference among our guests.
